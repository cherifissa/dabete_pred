```python
import numpy as np
import pandas as pd
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from keras.layers import Dense
from keras import Sequential
```

```python
# Charger le jeu de données
data = pd.read_csv('diabetes.csv', delimiter=',')
df = data.copy()
df.head(10)
```

```python
# Séparer les caractéristiques (X) et la variable cible (y)
# X = df.drop('Outcome', axis=1)
# y = df['Outcome']

# Diviser le jeu de données en ensembles d'entraînement et de test
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

X_train, X_test, y_train, y_test = train_test_split(df[['Pregnancies','Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']], df['Outcome'], test_size=0.33)
```

```python
# Standardiser les caractéristiques en utilisant StandardScaler
X_train_scaled = X_train.copy()
X_train_scaled = StandardScaler().fit_transform(X_train_scaled)

X_test_scaled = X_test.copy()
X_test_scaled = StandardScaler().fit_transform(X_test_scaled)
```

```python
# Construire le modèle
model = Sequential()
model.add(Dense(16, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compiler le modèle
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train_scaled, y_train, epochs=500)
```

```python
model.evaluate(X_test_scaled, y_test)
```

```python
# Faire des prédictions
predictions = model.predict(X_test_scaled)
```

```python
# Évaluer le modèle
loss, accuracy = model.evaluate(X_test_scaled, y_test)
```
